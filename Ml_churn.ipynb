```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Data Preparation
#### Drop insignificant columns
df=df.drop(["RowNumber","CustomerId","Surname"],axis=1) #axis=1 means column
```

```python
df.shape
```

```python
#### Missing values
df.isnull().sum()
```

```python
### Encoding
df.info()
```

```python
df=pd.get_dummies(df,columns=["Gender","Geography"])
```

```python
df.sample(10)
```

```python
df.shape
```

```python
### mv,encoded,dropped insignificant columns
```

```python
### Modelling
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #model name(classifier)
Y=df[["Exited"]]
X=df.drop("Exited",axis=1)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train)
```

```python
model_lr.score(X_train,Y_train)
```

```python
model_lr.score(X_test,Y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn=KNeighborsClassifier()
model_knn.fit(X_train,Y_train)
print(model_knn.score(X_train,Y_train))
print(model_knn.score(X_test,Y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier(max_depth=5)
model_dt.fit(X_train,Y_train)
print(model_dt.score(X_train,Y_train))
print(model_dt.score(X_test,Y_test))
```
