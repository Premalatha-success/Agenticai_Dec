## Load Data

### Subtask:
Load the dataset from the URL 'https://raw.githubusercontent.com/Premalatha-success/Datasets/refs/heads/main/CardioGoodFitness-1.csv' into a pandas DataFrame.


**Reasoning**:
To load the dataset, I need to first import the pandas library and then use its `read_csv` function to fetch the data from the provided URL and store it in a DataFrame.



```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Premalatha-success/Datasets/refs/heads/main/CardioGoodFitness-1.csv'
df = pd.read_csv(url)

print("Dataset loaded successfully. Displaying the first 5 rows:")
print(df.head())
```

## Initial Data Exploration

### Subtask:
Display the first few rows of the DataFrame, check its information (data types, non-null counts), and describe basic statistics for numerical columns.


**Reasoning**:
To perform the initial data exploration, I will display the first few rows, check the DataFrame's information, and generate descriptive statistics for numerical columns using pandas methods.



```python
print("Displaying the first 5 rows of the DataFrame:")
print(df.head())

print("\nConcise summary of the DataFrame (info()):")
df.info()

print("\nDescriptive statistics for numerical columns (describe()):")
print(df.describe())
```

## Generate Histograms

### Subtask:
Generate histograms for all numerical columns to visualize their distributions. Ensure appropriate legends are provided.


**Reasoning**:
To generate histograms for numerical columns, I need to import the necessary plotting libraries, identify the numerical columns, and then iterate through them to create and display each histogram with appropriate titles and labels.



```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Generating histograms for numerical columns...")

numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

print("Histograms generated successfully.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Generating box plots...")

# Define the combinations for box plots
plot_combinations = [
    ('Age', 'Product'),
    ('Income', 'Product'),
    ('Miles', 'Product'),
    ('Age', 'Gender'),
    ('Fitness', 'Gender'),
    ('Income', 'MaritalStatus')
]

for numerical_col, categorical_col in plot_combinations:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x=categorical_col, y=numerical_col, palette='viridis', hue=categorical_col, legend=False)
    plt.title(f'Distribution of {numerical_col} by {categorical_col}')
    plt.xlabel(categorical_col)
    plt.ylabel(numerical_col)
    plt.grid(axis='y', alpha=0.75)
    plt.show()

print("Box plots generated successfully.")
```

## Generate Correlation Heatmap

### Subtask:
Calculate the correlation matrix for numerical features and visualize it using a heatmap to understand relationships between variables. Ensure appropriate legends are provided.


**Reasoning**:
To generate the correlation heatmap, I will first select all numerical columns from the DataFrame, calculate their correlation matrix, and then use `seaborn.heatmap` to visualize this matrix with annotations and a clear title.



```python
print("Calculating correlation matrix and generating heatmap...")

numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

print("Correlation heatmap generated successfully.")
```
